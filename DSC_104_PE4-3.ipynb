```r
for (t in 1:40) {
    row <- new_Gervals$text[t]
    row <- tolower(row)
    row <- removeWords(row, stopwords("en"))
    row <- str_replace_all(row, "[[:punct:]]", "")
    row <- str_replace_all(row, "[0-9]+", " ")
    row <- str_replace_all(row, "al", " ")
    row <- str_replace_all(row, "\\s+", " ")
    new_row$answer[1] <- t
    new_row$text[1] <- row
    new_Gervals_clean <- rbind(new_Gervals_clean, new_row)
}
```
