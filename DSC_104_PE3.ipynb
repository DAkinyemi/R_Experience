```r
dummy_type <- as.data.frame(dummy.code(new_data$type))
```

```r
dummy_country <- as.data.frame(dummy.code(new_data$birthplace_country))
```

```r
names(dummy_type)
```

```r
names(dummy_country)
```

```r
new_data$Aircraft <- dummy_type$Aircraft
new_data$Bakery <- dummy_type$Bakery
new_data$BatesFabric <- dummy_type$BatesFabric
new_data$Bleachery <- dummy_type$Bleachery
new_data$Brick <- dummy_type$Brick
new_data$DrugandSundrystore <- dummy_type$DrugandSundrystore
new_data$Hospital <- dummy_type$Hospital
new_data$MachineShop <- dummy_type$MachineShop
new_data$Shoe <- dummy_type$Shoe
new_data$Canada <- dummy_country$Canada
new_data$Greece <- dummy_country$Greece
new_data$USA <- dummy_country$USA
new_data$"Store (died and cut leather)" <- dummy_type$"Store (died and cut leather)"
```

```r
new_data
```

```r
new_data_model <- lm(start_y ~ Aircraft + Bakery + Bleachery + Brick + Hospital + Shoe + Canada + Greece + USA, data = new_data)
```

```r
summary(new_data_model)
```

```r
new_data_model2 <- lm(start_y ~ Aircraft + Bakery + Bleachery +  Hospital + Shoe + Canada + Greece + USA , data = new_data)
```

```r
summary(new_data_model2)
```
