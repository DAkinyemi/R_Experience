```r
png(file="total_matrix.png", width=2500, height=2500, res=300)
plot.igraph(graph1, layout=layout.fruchterman.reingold, 
            edge.width = .25,
            vertex.size = 4,
            vertex.label=V(graph1)$name, 
            vertex.label.cex = .25,
            asp = 0
            )
dev.off()
```
